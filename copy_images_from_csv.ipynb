```python
import os
import pandas as pd
import shutil
```

```python
"""
Este script lê um arquivo CSV contendo informações de pastas e imagens, e tenta copiar
cada imagem da pasta de origem para a pasta de destino. Se uma imagem não for encontrada,
ela é registrada em uma lista de imagens não existentes. O script também trata e reporta
erros durante o processo de cópia.

Utilização
-- organizar as masks de IA


@Author Matheus Bury 
09/07/2024
version: 0
"""
```

```python
df = pd.read_csv(r"C:\Users\Matheus\Downloads\sqllab_untitled_query_54_20240802T020022.csv")

root = r"F:\1\pasta do model twin"
copy_root = r"F:\1\dataset"
```

```python
# Replace 'nan' string with actual NaN values
df.replace('nan', pd.NA, inplace=True)

# Fill NaN values with a placeholder string, for example 'unknown'
df['folder'].fillna('unknown', inplace=True)
df['image'].fillna('unknown', inplace=True)

# Ensure all values are strings
df['folder'] = df['folder'].astype(str)
df['image'] = df['image'].astype(str)

image_non_existe = []

for folder,image in zip(df['folder'],df['image']):
    #print(f'Pasta {folder}/{image}')
    
    path_folder_image = os.path.join(root,folder,image)
    path_copy_root = os.path.join(copy_root,image)
    
    
    
    
    try:
        print(f'{path_copy_root} --- movendo para --> {path_folder_image}')
        os.makedirs(os.path.join(root,folder),exist_ok=True)
        shutil.move(path_copy_root,path_folder_image)
        
    except FileNotFoundError:
        
        #print(f'File not found: {path_copy_root}')
        image_non_existe.append({'folder': folder, 'image': image})
        pass
        
    except OSError as e:
        #print(f'Error copying {path_folder_image} to {path_copy_root}: {e}')
        pass
#print("Non-existent images:")
for item in image_non_existe:
    print(f"Folder: {item['folder']}, Image: {item['image']}")
```
